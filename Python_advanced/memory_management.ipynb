Weak References


```python
import weakref

class A: pass

obj = A()
w = weakref.ref(obj)

print(w())   # returns object
del obj
print(w())   # None (object destroyed)
#w = weakref.ref(obj)
```

```python
from abc import ABC, abstractmethod

class Observer(ABC):
    @abstractmethod
    def update(self, message):
        pass

class ConsoleObserver(Observer):
    def update(self, message):
        print(f"[Console] {message}")

import weakref

class Subject:
    def __init__(self):
        self._observers = []   # stores weak references

    def attach(self, observer):
        self._observers.append(weakref.ref(observer))

    def notify(self, message):
        for w in self._observers:
            obs = w()
            if obs:                  # object still alive
                obs.update(message)
subject = Subject()

obs1 = ConsoleObserver()
subject.attach(obs1)

subject.notify("Training started")

del obs1        # observer removed automatically

subject.notify("Training finished")
```
