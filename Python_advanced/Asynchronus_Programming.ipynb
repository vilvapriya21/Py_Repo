```python
import asyncio

# 1. Define async function
async def my_async_function():
    print("Starting...")
    await asyncio.sleep(1)  # Simulate async operation
    print("Done!")
    return "result"

# 2. Run async function
result = asyncio.run(my_async_function())
print(result)

# 3. Multiple async tasks
async def main():
    # Run in parallel
    results = await asyncio.gather(
        my_async_function(),
        my_async_function(),
        my_async_function()
    )
    return results

asyncio.run(main())
```

```python
async def loader(dataset):
    print("Loading ",dataset)
    await asyncio.sleep(2)
    print("Loaded ",dataset)
    return f"Done {dataset}"

async def main():
    results = await asyncio.gather(
        loader('A'),
        loader('B'),
        loader('C')
    )

asyncio.run(main())
```

```python
import asyncio
import time

async def async_task():
    print("Async start")
    await asyncio.sleep(2)
    print("Async end")

def sync_task():
    print("Sync start")
    time.sleep(2)
    print("Sync end")

async def main():
    start = time.time()
    await asyncio.gather(async_task(), async_task())
    print("Async total:", time.time() - start)

start = time.time()
sync_task()
sync_task()
print("Sync total:", time.time() - start)

asyncio.run(main())
```
