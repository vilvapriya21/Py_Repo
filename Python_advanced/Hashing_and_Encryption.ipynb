```python
import hashlib

password="mypassword123" # if letters change hash also changes -->avalanche effect

hashed=hashlib.sha256(password.encode()).hexdigest()

print(hashed)
```

Salting

```python
import hashlib
import os

password = "mypassword123"
salt = os.urandom(16)  # random 16 bytes

hashed = hashlib.sha256(salt + password.encode()).hexdigest()

print("Salt:", salt)
print("Hash:", hashed)
```

```python
import hashlib

text = "sensitive_data"

# Different algorithms
algorithms = {
    'MD5': hashlib.md5,           # ❌ Weak - don't use for security
    'SHA-1': hashlib.sha1,        # ❌ Weak - don't use for security
    'SHA-256': hashlib.sha256,    # ✅ Good for most uses
    'SHA-512': hashlib.sha512,    # ✅ More secure, longer hash
}

print("Comparing hash algorithms:")
print("="*70)

for name, algorithm in algorithms.items():
    hash_value = algorithm(text.encode()).hexdigest()
    print(f"{name:10s}: {hash_value}")
```

| Algorithm | Secure?  | Use case       |
| --------- | -------- | -------------- |
| `md5`     | ❌ Broken | Legacy only    |
| `sha1`    | ❌ Broken | Legacy only    |
| `sha256`  | ✅ Okay   | Data integrity |
| `sha512`  | ✅ Okay   | Data integrity |
| `bcrypt`  | ✅ Best   | Passwords      |
| `argon2`  | ✅ Best   | Passwords      |

```python
import bcrypt

password = b"mypassword123"

hashed = bcrypt.hashpw(password, bcrypt.gensalt()) #salt included, slow by design and secure

print(hashed)

bcrypt.checkpw(b"mypassword123", hashed)  # True
bcrypt.checkpw(b"wrongpass", hashed)      # False
```

```python
from cryptography.fernet import Fernet

#generate key
key=Fernet.generate_key()
print(key)
cipher=Fernet(key)
print(cipher)

data=b"My data" #b-->bytes 

#encryption
encrypted=cipher.encrypt(data) #data -->bytes
print(encrypted)

#decrypt
decrypted=cipher.decrypt(encrypted)
print(decrypted)
```
