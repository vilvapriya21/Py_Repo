```python
df=pd.read_csv("Data/Zomato-data-.csv")
```

```python
df.head()     # first 5 rows
df.tail()     # last 5 rows
df.shape     # (rows, columns)
df.columns   # column names
df.info()    # data types & nulls
```

```python
df['name'] #one column

df[['name','online_order']] #multiple
```

loc (Label-based): Selects data by the names of the rows and columns.

iloc (Integer-positional): Selects data by its numerical position (starting from 0), regardless of what the labels are. 

```python
df.iloc[0] #first row

df.iloc[0:5]

print(df.loc[20]) # row by index label
print(df.iloc[20])
```

Filtering Data

```python
df[df['online_order']=='Yes']

df[(df['online_order']=='Yes') & (df['book_table']=='Yes')]

df[(df['listed_in(type)']=='Buffet') & (df['approx_cost(for two people)']<1000)]
```

df["column"] → Series → .sum() → Single number

df[["column"]] → DataFrame → .sum() → Series with column names

DataFrame .sum() default → Sums down each column (axis=0)

Use single bracket for quick numeric results

Use double bracket when you need column names in output

```python
df.isnull()
df.isnull().sum() 
print(df["listed_in(type)"].isnull().sum())
print(df[["listed_in(type)"]].isnull().sum())
```

```python
df.dropna()        # remove rows
df.fillna(0)       # replace with 0
```

```python
df["listed_in(type)"].value_counts()
df["listed_in(type)"].value_counts(dropna=False)
```

```python
df.index
df_with_index = df.set_index('name')
print(df_with_index)
df_reset = df.reset_index()
print(df_reset)
```
