def __new__(cls, name, bases, dct):

| Parameter | Meaning                         |
| --------- | ------------------------------- |
| `cls`     | Metaclass itself                |
| `name`    | Class name (`User`)             |
| `bases`   | Parent classes                  |
| `dct`     | Class body (methods, variables) |


```python
class MyMeta(type):
    def __new__(cls, name, bases, dct):
        print("Creating class:", name)
        return super().__new__(cls, name, bases, dct)

class User(metaclass=MyMeta):
    pass
```

```python
class EnforceID(type):
    def __new__(cls, name, bases, dct):
        if "id" not in dct:
            raise TypeError(f"{name} must have an 'id' attribute")
        return super().__new__(cls, name, bases, dct)

class Employee(metaclass=EnforceID):
    # id = 101
    pass
```
