Function    Works on	    Meaning

split	    1D or rows	    General split

hsplit	    columns	        Column-wise split

vsplit	    rows	        Row-wise split

```python
import numpy as np

# ------------------------------------
# 1️⃣ ORIGINAL DATA (1D array)
# ------------------------------------
a = np.array([10, 20, 30, 40])

print("Original array:", a)
print()

# ------------------------------------
# 2️⃣ SPLIT INTO EQUAL PARTS
# ------------------------------------
equal_split = np.split(a, 2)

print("Equal split into 2 parts:")
print(equal_split)
print()

# ------------------------------------
# 3️⃣ POSITION-BASED SPLIT
# Cut BEFORE index 1 and BEFORE index 3
# ------------------------------------
position_split = np.split(a, [1, 3])

print("Position-based split at indexes [1, 3]:")
print(position_split)
print()

# ============================================================
# NOW 2D ARRAY (TABLE-LIKE DATA)
# ============================================================

# ------------------------------------
# 4️⃣ 2D ARRAY (ROWS × COLUMNS)
# ------------------------------------
A = np.array([
    [1, 2, 3, 4],
    [5, 6, 7, 8]
])

print("2D Array:")
print(A)
print()

# ------------------------------------
# 5️⃣ HORIZONTAL SPLIT (COLUMNS)
# Equal split into 2 parts
# ------------------------------------
h_equal = np.hsplit(A, 2)

print("hsplit into 2 equal column parts:")
for part in h_equal:
    print(part)
print()

# ------------------------------------
# 6️⃣ POSITION-BASED HSPLIT
# Cut BEFORE column 1 and BEFORE column 3
# ------------------------------------
h_position = np.hsplit(A, [1, 3])

print("hsplit with positions [1, 3]:")
for part in h_position:
    print(part)
print()

# ------------------------------------
# 7️⃣ REAL-LIFE STYLE DATA
# ID | Age | Salary | Label
# ------------------------------------
data = np.array([
    [101, 25, 50000, 1],
    [102, 30, 60000, 0]
])

print("Real-life data:")
print(data)
print()

# ------------------------------------
# 8️⃣ SPLIT REAL DATA USING POSITIONS
# ------------------------------------
id_col, features, label = np.hsplit(data, [1, 3])

print("ID column:")
print(id_col)
print()

print("Features (Age, Salary):")
print(features)
print()

print("Label:")
print(label)
```
