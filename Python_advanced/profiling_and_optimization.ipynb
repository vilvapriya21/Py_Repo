timeit

```python
import timeit

def method1():
    li=[]
    for i in range(1000):
        li.append(i*i)
    return li

def method2():
    return [i for i in range(1000)]

print(timeit.timeit(method1,number=1000))
print(timeit.timeit(method2,number=1000))
```

```python
#a timeit usecase
import numpy as np
import timeit

# You're choosing between two ways to normalize data

def normalize_python_loop(data):
    """Using plain Python"""
    mean = sum(data) / len(data)
    return [(x - mean) for x in data]

def normalize_numpy(data):
    """Using NumPy"""
    return data - np.mean(data)

# Test data
data = list(range(100000))
data_np = np.array(data)

# Time them
time_loop = timeit.timeit(lambda: normalize_python_loop(data), number=100)
time_numpy = timeit.timeit(lambda: normalize_numpy(data_np), number=100)

print(f"Python loop: {time_loop:.4f}s")
print(f"NumPy: {time_numpy:.4f}s")
print(f"NumPy is {time_loop/time_numpy:.0f}x faster!")

# Output: NumPy is usually 50-100x faster!
```

```python
import cProfile
import time

def load_data():
    """Simulate loading data"""
    time.sleep(1)  # Pretend this takes 1 second
    return [1, 2, 3, 4, 5]

def preprocess_data(data):
    """Simulate preprocessing"""
    time.sleep(2)  # Pretend this takes 2 seconds
    return [x * 2 for x in data]

def train_model(data):
    """Simulate training"""
    time.sleep(0.5)  # Pretend this takes 0.5 seconds
    return sum(data)

def main():
    data = load_data()
    processed = preprocess_data(data)
    result = train_model(processed)
    return result

# Profile the code
cProfile.run('main()')
```
