```python
from abc import abstractmethod,ABC
```

```python
class Model(ABC):
    @abstractmethod #must be implemented in child classes 
    def train(self):
        pass

class MLModel(Model):
    def train(self):
        pass

#m=Model() #error if inherited from ABC 
```

```python
class ABCClass(ABC):
    x=11
    @abstractmethod
    def train(self,name):
        print("Hii",name)
class ABCChild(ABCClass):
    def train(self,name):
        print(name,"Hiiiii")

a=ABCChild()
a.train("me")
```

```python
class ABCClass(ABC):
    x = 11

    @abstractmethod
    def train(self, name):
        print("Hi", name)   # default behavior

class ABCChild(ABCClass):
    def train(self, name):
        super().train(name)   # reuse parent logic
        print(name, "Hiiiii")

ABCChild().train("me")
```

```python
class Parent(ABC):
    @property
    @abstractmethod
    def source_name(self):
        pass
    @source_name.setter # enforcing setter too
    @abstractmethod 
    def source_name(self, value):
        pass
class Child(Parent):

    def __init__(self):
        self._source_name = "My property"

    @property
    def source_name(self):
        return self._source_name

    @source_name.setter
    def source_name(self, name):
        self._source_name = name

    def display(self):
        print(self.source_name)

Child().display()
```

```python
class A(ABC):
    @abstractmethod
    def f(self): 
        print("f")

class B(A):
    def f(self):
        pass
    @abstractmethod
    def g(self): 
        print("g")

class C(B):
    def f(self): 
        super().f()
    def g(self): 
        super().g()
c=C()
c.f()
c.g()
```
