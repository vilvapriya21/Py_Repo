```python
import sqlite3

conn = sqlite3.connect("Databases/students.db")  # creates file if not exists
cursor = conn.cursor()

cursor.execute("""
    CREATE TABLE IF NOT EXISTS students (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        name TEXT,
        marks INTEGER
    )
""")

conn.commit()

cursor.execute(
    "INSERT INTO students (name, marks) VALUES (?, ?)",
    ("sammy", 95)
)
conn.commit()

data = [
    ("Priya", 90),
    ("Koma", 85),
    ("Janu", 88)
]

cursor.executemany(
    "INSERT INTO students (name, marks) VALUES (?, ?)",
    data
)
conn.commit()

cursor.execute("SELECT * FROM students")
rows = cursor.fetchall()

for r in rows:
    print(r)
```
