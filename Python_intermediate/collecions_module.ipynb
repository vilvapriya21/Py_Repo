```python
from collections import Counter

sentence='abcabsdakdjffffvd'
print(Counter(sentence))

words="Hi helloo hi hoi Hi helloo".split()
print(Counter(words))

numbers=[1,2,2,3,4,2,3,4,4,2,9,3,4]
frequency_dict=Counter(numbers)

print(frequency_dict)
print(frequency_dict.most_common(3))

frequency_dict.update([1,2,3,4,5,6,7,8,9])
print(frequency_dict)
```

```python
from collections import OrderedDict

od=OrderedDict()
od['a']=1
od['b']=2
od['c']=3

print(od)

od.move_to_end('a')
od.move_to_end('a', last=False)
od.move_to_end('b', last=False)
print(od)

print(od.popitem(last=True))
print(od.popitem(last=False))
```

```python
#default dict
from collections import defaultdict

d = defaultdict(int)
d["vilva"] += 1
d["2nd val"]+=3
print(d)

students=[
    ("vilva", "CSE"),
    ("koma", "ECE"),
    ("priya", "CSE"),
    ("keerthi", "ECE")
]

grouped_default_dict=defaultdict(list)

for name,dept in students:
    grouped_default_dict[dept].append(name[0])
print(grouped_default_dict)

words = ["apple","apple", "ant", "ball", "bat", "cat", "car"]
#task1
frequency=defaultdict(int)
for i in words:
    frequency[i]+=1
print(frequency)

#task2
first_words=defaultdict(list)
for i in words:
    first_words[i[0]].append(i)
print(first_words)
```

```python
#dequeue
from collections import deque

dq = deque([10, 20, 30])

dq.append(40)
print(dq)
dq.appendleft(5)
print(dq)
dq.pop()
print(dq)
dq.popleft()
print(dq)
dq.rotate(1)  # last element moves to front
print(dq)
dq.rotate(-1) # first element moves to end
print(dq)

```
