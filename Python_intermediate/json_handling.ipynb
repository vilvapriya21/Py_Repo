```python
json_str = '{"name": "Vilva", "age": 21, "skills": ["python", "ml"]}'

import json

data = json.loads(json_str)#json to python  object
print(data)
print(type(data))
```

```python
import json

d = {"course": "Python", "level": "beginner", "price": 0}
js = json.dumps(d, indent=4) #python object to json string
print(js)
print(type(js))
```

```python
import json
s={"name": "Priya", "age": 22, "hobbies": ["music", "coding"], "active": True}
js=json.dumps(s,indent=4)
print(js)
```

```python
import json

try:
    with open('json_file.json', 'r') as j:
        data=json.load(j) #loading from file
        print(data)
except FileNotFoundError:
    print("FileNotFoundError")
```

```python
import json
try:
    with open('json_file.json', 'w') as j:
        user={'name': 'Vilva Priya', 'age': 22, 'hobbies': ['music', 'coding']}
        json.dump(user,j,indent=4) #writing to file
except FileNotFoundError:
    print("FileNotFoundError")
```

```python
import json
with open('product.json','w') as j:
    product = {
       "id": 101,
       "name": "Laptop",
       "price": 55000,
       "avail": True
    }
    json.dump(product,j,indent=4)
with open('product.json','r') as j:
    product=json.load(j)
    print(product)
```
