```python
def my_decorator(func):
    def wrapper():
        print("Before function runs")
        func()
        print("After function runs")
    return wrapper

@my_decorator
def greet():
    print("Hello!")

greet()
```

```python
def log_decorator(func):
    def wrapper():
        print("Function started")
        func()
        print("Function finished")
    return wrapper

@log_decorator
def display():
    print("Hello user!")

display()
```

wraps to preserve the actual functions identity 

```python
from functools import wraps
def deco(func):
    @wraps(func)
    def wrapper():
        print('Hi')
        func()
        print('bye')
    return wrapper
@deco
def greetings():
    print('welcome')
greetings()
print(greetings.__name__)
```
