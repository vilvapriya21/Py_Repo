```python
import zipfile

files = ["books.csv", "practice.log", "practice_zip.zip"]

with zipfile.ZipFile("zip_practice.zip", "w", zipfile.ZIP_DEFLATED) as z:
    for f in files:
        z.write(f)
```

```python
import zipfile
import os

folder = "practice_dir"

with zipfile.ZipFile("zip_practice.zip", "w", zipfile.ZIP_DEFLATED) as z:
    for root, dirs, files in os.walk(folder):
        for file in files:
            full_path = os.path.join(root, file)
            arcname = os.path.relpath(full_path, folder)
            z.write(full_path, arcname)
```

```python
with zipfile.ZipFile("zip_practice.zip", "r") as z:
    print(z.namelist())
```

```python
with zipfile.ZipFile("zip_practice.zip", "r") as z:
    for info in z.infolist():
        print(info.filename, info.file_size, info.compress_size)
```
