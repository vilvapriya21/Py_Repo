```python
class Animal:
    def __init__(self):
        print("Animal")
    def eat(self):
        print("Animal is Eating")
class Dog(Animal):
    def __init__(self):
        super().__init__()
        print("Dog")
    def eat(self):
        print("Dog is Eating")
Dog().eat()
Animal().eat()
```

```python
class Animal:
    def speak(self): return "sound"

class Dog(Animal):
    def speak(self): return "woof"

class Cat(Animal):
    def speak(self): return "meow"

for a in (Dog(), Cat()):
    print(a.speak())
```

```python
#multiple 
#method resolution order(mro)
class Father:
    def skill1(self):
        print("Father's skill: Gardening")
    def skill2(self):
        print("Father's skill: cooking")

class Mother:
    def skill2(self):
        print("Mother's skill: Cooking")

class Child(Father, Mother):
    pass

c = Child()
c.skill1()
c.skill2()
```

```python
class A:
    def show(self):
        print("Class A")

class B(A):
    def show(self):
        print("Class B")

class C(A):
    def show(self):
        print("Class C")

class D(B, C):
    pass

d = D()
print(D.mro())
d.show()
```

```python
class A:
    def do(self): print("A")
class B(A):
    def do(self):
        print("B start")
        super().do()
        print("B end")
class C(A):
    def do(self):
        print("C start")
        super().do()
        print("C end")
class D(B, C):
    def do(self):
        print("D start")
        super().do()
        print("D end")

D().do()
```

```python
#polymorphism - duck typing
class Duck:
   def sound(self):
      return "Quack, quack!"

class AnotherBird:
   def sound(self):
      return "I'm similar to a duck!"

def makeSound(duck):
   print(duck.sound())

# creating instances
duck = Duck()
anotherBird = AnotherBird()
# calling methods
makeSound(duck)
makeSound(anotherBird)
```

```python
#operator overloading
class Vector:
   def __init__(self, a, b):
      self.a = a
      self.b = b

   def __str__(self):
      return 'Vector (%d, %d)' % (self.a, self.b)

   def __add__(self,other):
      return Vector(self.a + other.a, self.b + other.b)

v1 = Vector(2,10)
v2 = Vector(5,-2)
print (v1 + v2)
```

```python
#method overloading
def add(*nums):
   return sum(nums)

# Call the function with different number of parameters
result1 = add(10, 25)
result2 = add(10, 25, 35)

print(result1)
print(result2)
```

```python
class Employee:
   def __init__(self,nm, sal):
      self.name=nm
      self.salary=sal
   def getName(self):
      return self.name
   def getSalary(self):
      return self.salary

class SalesOfficer(Employee):
   def __init__(self,nm, sal, inc):
      super().__init__(nm,sal)
      self.incnt=inc
   def getSalary(self):
      return self.salary+self.incnt

e1=Employee("Rajesh", 9000)
print ("Total salary for {} is Rs {}".format(e1.getName(),e1.getSalary()))
s1=SalesOfficer('Kiran', 10000, 1000)
print ("Total salary for {} is Rs {}".format(s1.getName(),s1.getSalary()))
```
