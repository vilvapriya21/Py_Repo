```python
import csv

rows = [
    ["name","marks"],
    ["vilva",85],
    ["priya",90]
]
try:
    with open('students.csv','w',newline="") as csv_file: #newline='' to avoid lines between data
        writer=csv.writer(csv_file)
        writer.writerows(rows)
except FileNotFoundError:
    print('File not found')
```

```python
import csv

try:
    with open('students.csv','r') as csv_file:
        reader=csv.reader(csv_file)
        for row in reader:
            print(row)
except FileNotFoundError:
    print('File not found')

try:
    rows=[
        ['name', 'marks'],
        ['vilva', '85'],
        ['priya', '90']
    ]
    with open('books.csv','w',newline='') as csv_file:
        writer=csv.writer(csv_file)
        writer.writerows(rows)
except FileNotFoundError:
    print('File not found')
```

```python
import csv

try:
    with open('students.csv','r') as csv_file:
        dict_reader=csv.DictReader(csv_file)
        for i in dict_reader:
            print(i)
except FileNotFoundError:
    print("File not present")
```

```python
import csv

try:
    data=[
        {'book':'book1','price':111},
        {'book':'book2','price':222},
    ]
    with open('books.csv','w',newline='') as csv_file:
        dict_writer=csv.DictWriter(csv_file,fieldnames=['book','price'])
        dict_writer.writeheader()
        dict_writer.writerows(data)
except FileNotFoundError:
    print("File not present")
```
