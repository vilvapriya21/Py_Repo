```python
from multiprocessing import Process

def task(name):
    print("task is going on ",name)

if __name__=="__main__":
    p=Process(target=task,args=("task1",))

    p.start()

    p.join()

    print("Finished")
```

```python
from multiprocessing import Process
import time

def work(n):
    print(f"Process {n} started")
    time.sleep(2)
    print(f"Process {n} finished")

processes = []

for i in range(1, 4):
    p = Process(target=work, args=(i,))
    p.start()
    processes.append(p)

for p in processes:
    p.join()

print("All processes done")
```

Pool

Normal map → single CPU 
Pool map → multiple CPUs
p.map(square, [1, 2, 3, 4, 5])

apply → one input
p.apply(square, args=(10,))

```python
#Pool example 
from multiprocessing import Pool

def square(x):
    return x * x

if __name__ == "__main__":
    with Pool(processes=3) as p:
        result = p.map(square, [1, 2, 3, 4, 5])

    print(result)
```

Pool.starmap (multiple arguments)

```python
def add(a, b):
    return a + b

if __name__ == "__main__":
    with Pool(2) as p:
        result = p.starmap(add, [(1,2), (3,4), (5,6)])

    print(result)
```

```python
#practice tasks
from multiprocessing import Pool

def cube(x):
    return x * x * x

if __name__ == "__main__":
    with Pool(processes=3) as p:
        li=range(1,11)
        result = p.map(cube,li)

    print(result,type(result))
```
