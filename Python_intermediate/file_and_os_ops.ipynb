```python
from pathlib import Path

base = Path("Python_intermediate")  # Path object
sub = base / "logs"     # Use `/` to join paths
file = sub / "security.log"

print(file)                 # project/data/notes.txt
print(file.exists())        # Check if exists
print(file.is_file())       # Is it a file?
print(file.is_dir())        # Is it a directory?
print(file.absolute())      # Absolute path
print(file.suffix)          # File extension
print(file.stem)            # File name without extension
```

```python
import shutil
print(os.listdir("."))
# Copy file
shutil.copy("source.txt", "destination.txt")       # copies contents, preserves permission
shutil.copy2("source.txt", "destination.txt")  # also copies metadata like mtime
# Copy folder
shutil.copytree("practice_dir", "dest_folder") # raises if dst exists
# Python 3.8+: shutil.copytree("src","dst", dirs_exist_ok=True)

# Move file/folder
shutil.move("a/b/c/file.txt", "practice_dir/")

# Delete folder
shutil.rmtree("a")
```

```python
# get disk usage
total, used, free = shutil.disk_usage("/")

# make archive (zip/tar)
shutil.make_archive("archive_name", "zip", root_dir="folder_to_zip")
# creates archive_name.zip

# unpack
shutil.unpack_archive("archive_name.zip", extract_dir="outdir")
```

```python
import os, shutil
from pathlib import Path

# 1 - Setup
p = Path("practice_dir/data")
p.mkdir(parents=True, exist_ok=True)

# 2 - Create files
for name in ("a.txt","b.txt","c.log"):
    (p / name).write_text(f"hello from {name}\n")

# 3 - scandir listing
with os.scandir("practice_dir") as it:
    for entry in it:
        print("entry:", entry.name, "is_file?", entry.is_file())

# 4 - copy and move
shutil.copy2("practice_dir/data/a.txt", "practice_dir/a_copy.txt")
shutil.move("practice_dir/data/b.txt", "practice_dir/data/moved_b.txt")

# 5 - make archive
shutil.make_archive("practice_zip", "zip", root_dir="practice_dir")

# 6 - extract
shutil.unpack_archive("practice_zip.zip", extract_dir="practice_dir_unzip")
print("Done")
```
