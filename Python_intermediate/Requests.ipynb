```python
import requests

response = requests.get("https://jsonplaceholder.typicode.com/posts/1")

print(response.status_code)   # 200 means success
print(response.text)          # raw text
print(response.json())        # parsed JSON
```

```python
import requests

url="https://jsonplaceholder.typicode.com/posts"
params={"title":"my post"}

response=requests.get(url,params=params)
print(response.json())
```

```python
import requests

data={
    "title": "my post",
    "body": "Hello!",
    "userId": 5
}

url="https://jsonplaceholder.typicode.com/posts"

response = requests.post(url, json=data)
print(response.json())
```

```python
import requests
url="https://httpbin.org"
response = requests.get(url,timeout=1)

try:
    response.raise_for_status()   # If the request failed, this raises an exception immediately
    print("Success")
    # Proceed to process the response. For example:
    # data = response.json()

except requests.exceptions.HTTPError as e:
    print(f"Failed due to HTTP error: {e}")
    # The program catches the error here instead of crashing entirely
```

```python
import requests
try:
    import json

    url = "https://httpbin.org"

    headers = {
        "Authorization": "Bearer 123456",
        "Content-Type": "application/json"
    }

    # Send the GET request with the headers
    response = requests.get(url, headers=headers)

    # Check if the request was successful (200 OK)
    if response.status_code == 200:
        print(response)
        print("Request successful. Server response:")
        # Print the full JSON response received from the server
        print(json.dumps(response.json(), indent=4))
    else:
        print(f"Request failed with status code: {response.status_code}")
except Exception as e:
    print(f"Request failed with HTTP error: {e}")
```
