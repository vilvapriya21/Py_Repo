```python
#HOF that takes function example
def apply_twice(func, x):
    return func(func(x))

def add3(n): return n + 3

print(apply_twice(add3, 7))
```

```python
#HOF that returns function
def make_multiplier(k):
    def mul(x):
        return x * k
    return mul

double = make_multiplier(2)
triple = make_multiplier(3)
print(double(5))  # 10
print(triple(5))  # 15

def power(n):
    def calc(x):
        return x ** n
    return calc

square = power(2)
cube = power(3)

print(square(5))  # 25
print(cube(5))    # 125
```

```python
#practice
'''
Create a function math_operation(fn, x, y)
that takes a function and two numbers.'''
def math_operation(fn,x,y):
    return fn(x,y)
def add(a, b): return a + b
def mul(a, b): return a * b

print(math_operation(add, 5, 3)) # 8
print(math_operation(mul, 5, 3)) # 15
```

```python
#built in HOFs
nums = [1,2,3,4]
print(list(map(lambda x: x*x, nums)))

print(list(filter(lambda x: x%2==0, nums)))

words = ["apple","a","banana"]
print(sorted(words, key=len))

from functools import reduce
print(reduce(lambda a,b: a+b, nums))
```

```python
#map(function,iterable)
def upper_case(x):
    return x.upper()
words=['vilva','is','a','good','girl']
print(list(map(upper_case,words)))

nums = [1, 2, 3, 4]

def square(x):
    return x * x

result = map(square, nums)
print(result," ",list(result))

```

```python
#filter
nums = [1, 2, 3, 4, 5, 6]

def is_even(n):
    return n % 2 == 0

print(list(filter(is_even, nums)))

words = ["cat", "elephant", "dog", "lion", "hi", "python"]

print(list(filter(lambda word: len(word)>3, words)))
```

```python
#lambda
even=lambda x: x%2==0
print(even(99))

words = ["hi", "python", "cat"]
print(sorted(words, key=lambda w: len(w)))

data = [(1, 5), (2, 1), (3, 4)]
print(sorted(data, key=lambda x: x[1]))

nums = [1, 2, 3, 4, 5]
print(list(map(lambda x:x*x,nums)))
```
