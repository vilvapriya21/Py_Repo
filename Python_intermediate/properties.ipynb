```python
class Person:
    def __init__(self,age):
        self.age=age
    @property
    def age(self):
        return self.__age
    @age.setter
    def age(self,new_age):
        if new_age > 0:
            self.__age=new_age
        else:
            raise ValueError("Invalid age")
    @age.deleter
    def age(self):
        del self.__age
# person_1=Person(-11)
# print(person_1.age)
person_1=Person(11)
person_1.age=18
print(person_1.age)
# del person_1.age
# print(person_1.age)
```

```python
#readonly
class Person:
    def __init__(self, name):
        self._name = name

    @property
    def name(self):          # getter
        return self._name

p = Person("Priya")
print(p.name)               # works
# p.name = "X"             # raises AttributeError
```

```python
#getter only - for read only computed values
class Circle:
    def __init__(self, radius):
        self._r = radius

    @property
    def radius(self):
        return self._r

    @property
    def area(self):
        return 3.14159 * self._r * self._r
```
