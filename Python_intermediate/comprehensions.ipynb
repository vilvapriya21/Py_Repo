```python
names = ["vilva", "priya", "koma", "ab", "xy"]
print([x.upper() for x in names])

#if else must be before the for
li=[x if len(x) > 4 else "shorter" for x in names]
print(li)
```

```python
nums=[1,2,3,4,5]
li=[x**2 for x in nums[::-1]]
print(li)
```

```python
matrix = [[1,2,3],[4,5,6],[7,8,9]]
flat = [num for row in matrix for num in row]
print(flat)
```

```python
pairs = [(x, y) for x in [1,2] for y in [10,20]]
print(pairs)
```

```python
a = [[1, 2], [3, 4], [5, 6]]
li=[(x,y) for y in a for x in y]
print(li)
```

```python
s="hello world"
print([x for x in s if x in ['a','e','i','o','u']])

print([x**2 for x in range(1,30) if x%3==0])

li=[[1,2], [3,4], [5,6,7]]

print([y for x in li for y in x])

```

```python
#{key: value for item in iterable}
names = ["vilva", "priya", "koma"]

d = {i: name for i, name in enumerate(names)}
print(d)

for i,k in enumerate(names):
    print(k)
```

```python
prices = {"apple": 50, "banana": 20, "kiwi": 100}
d={k:i*0.7 for k,i in prices.items()}
print(d)
```

```python
marks = {"math": 95, "eng": 45, "sci": 87}#reverse and compare
#d={value:key if value>50 else key:value for key,value in marks.items()} can't have two k:v
d = {
    (value, key) if value > 50 else (key, value)
    for key, value in marks.items()
}
print(d)
```

```python
d = {"a": 1, "b": 2, "c": 3}
swapped = {v: k for k, v in d.items()}
print(swapped)
```

```python
table = {
    i: {j: i*j for j in range(1,6)}
    for i in range(1,4)
}

print(table)
```
