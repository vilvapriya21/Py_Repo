Each next():

Resumes function from where it paused

Runs until the next yield

Pauses again

```python
def numbers():
    yield 1
    yield 2
    yield 3

g = numbers()

print(next(g))  
print(next(g))  
print(next(g))  
```

```python
def batch_generator(data, batch_size):
    for i in range(0, len(data), batch_size):
        yield data[i:i+batch_size]

data = list(range(10))

for batch in batch_generator(data, 3):
    print(batch)
```

```python
def even_numbers(data):
    for i in data:
        if i%2==0:
            yield i
    
for x in even_numbers([1,2,3,4,5,6]):
    print(x)

'''
for loop internally does this

while True:
    try:
        x = next(gen)
        print(x)
    except StopIteration:
        break
'''
```

custom iterator

```python
class Counter:
    def __init__(self, limit):
        self.limit = limit
        self.current = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.current > self.limit:
            raise StopIteration
        value = self.current
        self.current += 1
        return value
    
c = Counter(3)

for x in c:
    print(x)
```
