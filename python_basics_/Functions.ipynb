```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def __add__(self, other):
        return Point(self.x + other.x, self.y + other.y)

p1 = Point(2, 3)
p2 = Point(4, 1)
p3 = p1 + p2  # calls __add__()
print(p3.x, p3.y)  # 6 4
```

```python
def factorial(num: int) -> int:
    if num == 0:
        return 1
    else:
        return num * factorial(num - 1)
print(factorial(5))  # 120
```
