```python
"""
This module defines the Person class and demonstrates how to create objects
and display their details. This file can be run directly as a script.
"""
class Person:
    """Represents a person with a name and age."""
    def __init__(self, name, age):
        """
        Initialize a new Person object.

        Args:
            name (str): The person's name.
            age (int or str): The person's age.
        """
        self.name = name
        self.age = age

    def show(self):
        """
        Print the person's details.

        Returns:
            None
        """
        print(f"name is {self.name} and age is {self.age}")


if __name__ == "__main__":
    # Example usage
    person_1 = Person('abc', '11')
    person_2 = Person('cde', '18')
    person_3 = Person('efg', '22')

    person_1.show()
    person_2.show()
    person_3.show()
```

```python
'''Types of methods demo'''
class Student:
    college_name="ABC College"
    def __init__(self,name,age):
        self.name=name
        self.age=age
    def show(self):
        print(f"name:{self.name}\nage:{self.age}")
    @classmethod
    def college_change(cls,new_college_name):
        cls.college_name=new_college_name
        print("College name changed succesfully")
    @staticmethod
    def is_pass(marks):
        if marks>35:
            return True
        return None
if __name__ == '__main__':
    student = Student('me', 21)
    student.show()
    student.college_change('Yoo College')
    print(student.is_pass(44))

```

```python
# Create a class BankAccount with:
#
# ✔️ instance method: deposit()
# ✔️ class method: change_bank_name()
# ✔️ static method: validate_amount()
class BankAccount:
    bank_name="SBI"
    def __init__(self,name,balance):
        self.name=name
        self.balance=balance
    def deposit(self,amount):
        self.validate_amount(amount)
        self.balance=self.balance+amount
    @classmethod
    def change_bank_name(cls):
        cls.bank_name="SBI"
    @staticmethod
    def validate_amount(amount):
        if amount<0:
            raise ValueError("Amount cannot be negative")
```
