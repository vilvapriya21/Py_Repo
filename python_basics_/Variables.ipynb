```python
def count_up_to(n):
    i = 1
    while i <= n:
        yield i     # Pause here and send value to caller
        i += 1

counter = count_up_to(2)
print(next(counter))  # 1
print(next(counter))  # 2
```

```python
x = 10
def modify():
    global x
    x = 20  # modifies global variable
modify()
print(x) 
```

```python
def outer():
    x = 10
    def inner():
        nonlocal x
        x = 20  # modifies 'x' in outer, not global
    inner()
    print(x)  # 20

outer()
```
