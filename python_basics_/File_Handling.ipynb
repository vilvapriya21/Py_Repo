```python
with open("file_task.txt", "w") as f:
    f.write("my name: 'my name'\n")
    f.write("age: 'my age'\n")

with open("file_task.txt",'r') as f:
    print(f.readline())

file=open("file_task.txt",'a')
file.writelines("AI ML is an advanced technology")

# Write numbers
with open("file4.txt", "w") as f:
    f.write("1\n2\n3\n4\n5")

# Read & sum numbers
s = 0
with open("file4.txt", "r") as f:
    for line in f:
        s += int(line.strip())

print("Sum =", s)
```

```python
# Step 1: Read from source file
with open("file.txt", "r") as src:
    content = src.read()

# Step 2: Write to destination file
with open("file_task.txt", "w") as dest:
    dest.write(content)

print("Content copied successfully!")
```
